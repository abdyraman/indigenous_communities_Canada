```python
# Import dependencies
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
from selenium import webdriver
from sklearn.model_selection import train_test_split
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Create a list to hold data from all pages
all_page_data = []

# Iterate through the band numbers (from 172 to 173 in this example)
for i in range(1, 781):
    url = f'https://fnp-ppn.aadnc-aandc.gc.ca/fnp/Main/Search/FNMain.aspx?BAND_NUMBER={i}&lang=eng'
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    # Extract the band information
    band_name_element = soup.find('span', {'id': 'plcMain_txtBandName'})
    band_number_element = soup.find('span', {'id': 'plcMain_txtBandNumber'})
    band_address_element = soup.find('span', {'id': 'plcMain_txtAddress'})
    band_phone_element = soup.find('span', {'id': 'plcMain_txtPhone'})
    band_fax_element = soup.find('span', {'id': 'plcMain_txtFax'})
    band_website_element = soup.find('a', {'id': 'plcMain_anchor1'})

    # Extract the band information
    group_name = band_name_element.text if band_name_element else None
    group_num = band_number_element.text if band_number_element else None
    group_add = band_address_element.text if band_address_element else None
    group_phn = band_phone_element.text if band_phone_element else None
    group_fax = band_fax_element.text if band_fax_element else None
    group_web = band_website_element['href'] if band_website_element else None

    # Organize the data into a dictionary
    item_data = {
        'Band Name': group_name,
        'Band Number': group_num,
        'Band Address': group_add,
        'Band Phone': group_phn,
        'Band Fax': group_fax,
        'Band Website': group_web
    }
    all_page_data.append(item_data)

```

```python
# Convert the dictionary to a DataFrame
df = pd.DataFrame(all_page_data)
```

```python
# Save the DataFrame to Excel
file_path = 'raw_data_indigenous_list.xlsx'
df.to_excel(file_path, index=False)
```

```python
cleaned_df = df.dropna(thresh=5)
cleaned_df = cleaned_df.reset_index(drop=True)
cleaned_df
```

```python
# Save the DataFrame to Excel
file_path = 'cleaned_data_indigenous_list.xlsx'
cleaned_df.to_excel(file_path, index=False)
```
